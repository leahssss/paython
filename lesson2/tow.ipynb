```python
 a= range(2010, 2021)
 b = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
 c = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
 d = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
```

```python
fig,axs = plt.subplots(2, 2)  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(a, b, 'tab:blue')
axs[0, 1].plot(a, c, 'tab:red')
axs[1, 0].plot(a, d,'tab:black')
axs[1, 1].plot(d, c)
for ax in axs.flat:
    ax.label_outer()
```
